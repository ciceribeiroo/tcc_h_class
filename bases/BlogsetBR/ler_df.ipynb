```python
ids_authors = authors['AuthorID'].tolist()
```

```python
posts.shape
```

```python
texts_by_author = []
for id in ids_authors:
    df_author = posts[posts[5]==id]
    texts = df_author[4].tolist()
    complete_text = " ".join(texts)
    texts_by_author.append(complete_text)
```

```python
len(texts_by_author)
```

```python
authors["Texts"] = texts_by_author
```

```python
df = authors.drop(columns=['Timestamp', 'Sobre qual(is) tópico(s) você mais escreve?', 'Utiliza outra rede social?', 'Qual sua atividade profissional?', 'Qual o endereço do seu blogspot?',
'Onde você mora?', 'Código Único',
       'Há quanto tempo você mantém esse blog?',
       'Qual o período de atualização do blog?',
       'Costuma deixar comentários nos blogs que segue?',
       'Com relação a autoria do conteúdo de seu blog:',
       'Quantidade média de acessos do seu blog por mês?',
       'Nivel de Escolaridade', "Gostaria que seu blog fosse publicado na lista deste estudo?"])
```

```python
import numpy as np
```

```python
df['Qual sua idade?'] = df['Qual sua idade?'].replace('', np.nan)
df['Qual seu sexo?'] = df['Qual seu sexo?'].replace('', np.nan)
df['Texts'] = df['Texts'].replace('', np.nan)
```

```python
def getAgeGroups(text):
    if text == '10 à 18 anos' or text == '19 à 25 anos':
        return 0
    elif text == '26 à 30 anos' or text == '30 à 40 anos':
        return 1
    else:
        return 2
```

```python
df['AgeClass'] = df['Qual sua idade?'].apply(getAgeGroups)
```
